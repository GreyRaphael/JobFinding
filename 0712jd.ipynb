```python
import requests
```

```python
HEADERS={
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:89.0) Gecko/20100101 Firefox/89.0',
}
```

```python
json_data={"pageSize":10,"pageIndex":0,"parameter":{"positionName":"","planIdList":[18],"jobDirectionCodeList":[],"workCityCodeList":[],"positionDeptList":[]}}
```

```python
response=requests.post('https://campus.jd.com/api/wx/position/page?type=present', json=json_data, headers=HEADERS).json()
```

```python
response['body']['items'][0]['publishId'], response['body']['items'][0]['qualification']
```

```python
job_list=[]
for pg in range(2):
    json_data={"pageSize":10,"pageIndex":pg,"parameter":{"positionName":"","planIdList":[18],"jobDirectionCodeList":[],"workCityCodeList":[],"positionDeptList":[]}}
    response=requests.post('https://campus.jd.com/api/wx/position/page?type=present', json=json_data, headers=HEADERS).json()
    for job in response['body']['items']:
        job_list.append((job['publishId'], job['qualification']))
```

```python
len(job_list)
```

```python
for jobId, job in job_list:
    if job.find('物理')!=-1 or job.find('理工') !=-1:
        url=f'https://campus.jd.com/#/details?id={jobId}'
        print(url)
```

```python
for jobId, job in job_list:
    print(jobId, '-'*50)
    print(job)
```
